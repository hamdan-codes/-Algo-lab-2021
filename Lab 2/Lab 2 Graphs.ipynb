```python
import matplotlib.pyplot as plt
import numpy as np
```

# 3rd

# best 

```python
x = np.array([5000, 10000, 15000, 20000, 25000, 30000])
y = [0.000001, 0.000002, 0.000002, 0.000002, 0.000003, 0.000003]
plt.title('Best')
plt.xlabel('Iterations')
plt.xlabel('Time taken')
plt.scatter(x, y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)
plt.show()
```

# avg

```python
x = np.array([5000, 10000, 15000, 20000, 25000, 30000])
y = [0.000004, 0.000013, 0.000021, 0.000027, 0.000040, 0.000040]
plt.title('Average')
plt.xlabel('Iterations')
plt.xlabel('Time taken')
plt.scatter(x, y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)
plt.show()
```
