```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1st 

```python
x = np.array([100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000]) 
y1 = [0.000051, 0.000040, 0.000049, 0.000031, 0.000053, 0.000024, 0.000035, 0.000048, 0.000041, 0.000038]
y2 = [0.000007, 0.000012, 0.000006, 0.000011, 0.000010, 0.000006, 0.000015, 0.000009, 0.000005, 0.000008]
plt.xlabel('Input size')
plt.ylabel('Time taken')

plt.scatter(x, y1)
m, b = np.polyfit(x, y1, 1)
plt.plot(x, m*x+b, label="Simple Algo")

plt.scatter(x, y2)
m, b = np.polyfit(x, y2, 1)
plt.plot(x, m*x+b, label="Optimised Algo")

plt.legend()
plt.show()
```

```python
x = np.array([5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000])
yasc = [0.000024, 0.000046, 0.000125, 0.000090, 0.000128, 0.000134, 0.000159, 0.000246, 0.000249, 0.000255]
ydesc = [0.041080, 0.162470, 0.483918, 0.669699, 1.067751, 1.451958, 1.996108, 2.591937, 3.527495, 4.071260]
yrand = [0.021093, 0.082199, 0.236109, 0.326679, 0.534011, 0.732054, 0.997238, 1.283326, 2.105550, 2.037928]

plt.xlabel('Input size')
plt.ylabel('Time taken')

plt.scatter(x, yasc)
m, b = np.polyfit(x, yasc, 1)
plt.plot(x, m*x+b, label="Array already in ascending")

plt.scatter(x, ydesc)
m, b = np.polyfit(x, ydesc, 1)
plt.plot(x, m*x+b, label="Array in Descending")

plt.scatter(x, yrand)
m, b = np.polyfit(x, yrand, 1)
plt.plot(x, m*x+b, label="Array in Random state")

plt.legend()
plt.show()
```
