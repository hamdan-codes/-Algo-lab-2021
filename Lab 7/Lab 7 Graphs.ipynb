```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1st 

```python
x = np.array([10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]) 
yb = [0.000000, 0.008000, 0.008000, 0.008000, 0.009000, 0.008000, 0.016000, 0.016000, 0.024000, 0.008000] 
ya = [0.007000, 0.009000, 0.008000, 0.016000, 0.024000, 0.024000, 0.024000, 0.032000, 0.016000, 0.016000] 
yw = [0.000000, 0.000000, 0.008000, 0.008000, 0.016000, 0.016000, 0.016000, 0.024000, 0.024000, 0.024000] 

plt.xlabel('Input size')
plt.ylabel('Time taken')

plt.scatter(x, yb)
m, b = np.polyfit(x, yb, 1)
plt.plot(x, m*x+b, label="Best Case")

plt.scatter(x, ya)
m, b = np.polyfit(x, ya, 1)
plt.plot(x, m*x+b, label="Avg Case")

plt.scatter(x, yw)
m, b = np.polyfit(x, yw, 1)
plt.plot(x, m*x+b, label="Worst Case")

plt.legend()
plt.show()
```
