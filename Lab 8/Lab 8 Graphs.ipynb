```python
x = np.array([10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]) 
yb = [0.016000, 0.004000, 0.008000, 0.0130000, 0.016000, 0.015000, 0.016000, 0.031000, 0.022000, 0.031000] 

plt.xlabel('Input size')
plt.ylabel('Time taken')

plt.scatter(x, yb)
m, b = np.polyfit(x, yb, 1)
plt.plot(x, m*x+b, label="Best Case")

plt.show()
```
