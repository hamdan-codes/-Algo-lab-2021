```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1st 

```python
x = np.array([10000, 15000, 20000, 25000, 30000, 35000, 40000]) 
yb = [0.001000, 0.002000, 0.002000, 0.003000, 0.004000, 0.005000, 0.005000]
ya = [0.193000, 0.419000, 0.740000, 1.156000, 1.658000, 2.251000, 2.964000]
yw = [0.266000, 0.592000, 1.033000, 1.602000, 2.312000, 3.146000, 4.130000]

plt.xlabel('Input size')
plt.ylabel('Time taken')

plt.scatter(x, yb)
m, b = np.polyfit(x, yb, 1)
plt.plot(x, m*x+b, label="Best Case")

plt.scatter(x, ya)
m, b = np.polyfit(x, ya, 1)
plt.plot(x, m*x+b, label="Avg Case")

plt.scatter(x, yw)
m, b = np.polyfit(x, yw, 1)
plt.plot(x, m*x+b, label="Worst Case")

plt.legend()
plt.show()
```
