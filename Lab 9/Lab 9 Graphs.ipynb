```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1st 

```python
x = np.array([10000, 20000, 30000, 40000, 50000]) 
yb = [0.000000, 0.001000,0.001000,0.001000,0.002000] 

plt.xlabel('Input size')
plt.ylabel('Time taken')

plt.scatter(x, yb)
m, b = np.polyfit(x, yb, 1)
plt.plot(x, m*x+b)

plt.show()
```

# 2nd

```python
x = np.array([10000, 20000, 30000, 40000, 50000])
yb = [0.000000, 0.000000,0.000000,0.000000,0.000000]
ya = [0.000000, 0.000000,0.000000,0.000000,0.000000]
yw = [0.000000, 0.000000,0.000000,0.000000,0.000000]

plt.xlabel('Input size')
plt.ylabel('Time taken')

plt.scatter(x, yb)
m, b = np.polyfit(x, yb, 1)
plt.plot(x, m*x+b, label="Best Case")

plt.scatter(x, ya)
m, b = np.polyfit(x, ya, 1)
plt.plot(x, m*x+b, label="Avg Case")

plt.scatter(x, yw)
m, b = np.polyfit(x, yw, 1)
plt.plot(x, m*x+b, label="Worst Case")

plt.legend()

plt.show()
```

```python
x = np.array([10000, 20000, 30000, 40000, 50000])
ymx = [0.000000, 0.000000,0.000000,0.000000,0.000000]
ymn = [0.000000, 0.000000,0.000000,0.000000,0.000000]

plt.xlabel('Input size')
plt.ylabel('Time taken')

plt.scatter(x, ymx)
m, b = np.polyfit(x, ymx, 1)
plt.plot(x, m*x+b, label="Max Case")

plt.scatter(x, ymn)
m, b = np.polyfit(x, ymn, 1)
plt.plot(x, m*x+b, label="Min Case")

plt.legend()
plt.show()
```
